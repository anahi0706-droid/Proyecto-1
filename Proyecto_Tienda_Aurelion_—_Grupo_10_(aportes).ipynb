```python
# ==================================================================================================
# üß† PROYECTO: Tienda Aurelion
# Sprint 2 ‚Äî An√°lisis Estad√≠stico Descriptivo
# Autor:
# Versi√≥n: 2.0 | Noviembre 2025
# Archivo: notebook_tienda_aurelion.ipynb
# ==================================================================================================

# ================================================================================================
# üì¶ FASE 1 ‚Äî CARGA Y EXPLORACI√ìN DE LOS DATOS
# ================================================================================================

import os
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import unicodedata # Added this import statement

pd.set_option('future.no_silent_downcasting', True) # Added to suppress FutureWarning

# ------------------------------------------------------------
# 1Ô∏è‚É£ MONTAJE DE GOOGLE DRIVE
# ------------------------------------------------------------
print("üîó Montando Google Drive...")
drive.mount('/content/drive', force_remount=True)

ruta_base = "/content/drive/MyDrive/Colab Notebooks/Data_Aurelion"
carpeta_salida = os.path.join(ruta_base, "Resultados")

os.makedirs(carpeta_salida, exist_ok=True)

# ------------------------------------------------------------
# 2Ô∏è‚É£ LECTURA DE ARCHIVOS
# ------------------------------------------------------------
# ============================================================
# üîç Verificaci√≥n autom√°tica de archivos requeridos
# Aporte agregado por: Anah√≠ Gonz√°lez
# ============================================================

import os

def verificar_archivos_necesarios(ruta, archivos):
    print("\nüîé Verificando archivos requeridos en la ruta:")
    print(f"   üìÅ {ruta}\n")

    # Mostrar contenido real de la carpeta
    if os.path.exists(ruta):
        contenido = os.listdir(ruta)
        print(f"üìÑ Contenido actual: {contenido}\n")
    else:
        print("‚ùå La carpeta no existe. Revisa la ruta en 'ruta_base'.")
        return False

    todo_ok = True

    for archivo in archivos:
        ruta_archivo = os.path.join(ruta, archivo)

        if os.path.exists(ruta_archivo):
            print(f"‚úîÔ∏è OK: {archivo} encontrado.")
        else:
            print(f"‚ùå ERROR: {archivo} no est√° en la carpeta. C√°rgalo para continuar.")
            todo_ok = False

    if todo_ok:
        print("\n‚úÖ Todos los archivos est√°n presentes. Puede continuar la ejecuci√≥n.\n")
    else:
        print("\n‚ö†Ô∏è Falta uno o m√°s archivos. La ejecuci√≥n se detendr√° para evitar errores.")

    return todo_ok


archivos_requeridos = [
    "clientes.xlsx",
    "productos.xlsx",
    "ventas.xlsx",
    "detalle_ventas.xlsx"
]

if not verificar_archivos_necesarios(ruta_base, archivos_requeridos):
    raise FileNotFoundError("‚ùå No se puede continuar: faltan archivos requeridos.")


print("\nüì• Leyendo archivos...")
clientes = pd.read_excel(os.path.join(ruta_base, "clientes.xlsx"))
productos = pd.read_excel(os.path.join(ruta_base, "productos.xlsx"))
ventas = pd.read_excel(os.path.join(ruta_base, "ventas.xlsx"))
detalle = pd.read_excel(os.path.join(ruta_base, "detalle_ventas.xlsx"))

print(f"Clientes: {clientes.shape}, Productos: {productos.shape}, Ventas: {ventas.shape}, Detalle: {detalle.shape}")

# ------------------------------------------------------------
# 3Ô∏è‚É£ EXPLORACI√ìN INICIAL
# ------------------------------------------------------------
print("\nüîé Explorando estructura de los datos...")
for nombre, df in zip(["Clientes","Productos","Ventas","Detalle de Ventas"], [clientes, productos, ventas, detalle]):
    print(f"\n--- {nombre} ---")
    print(df.info())
    print(df.head(3))
    print("-"*60)
# ============================================================
# Diagn√≥stico autom√°tico de columnas faltantes
# Aporte agregado por: Anah√≠ Gonz√°lez
# ============================================================

def diagnostico_columnas(df, columnas_esperadas, nombre_tabla):
    print(f"\nüìã Diagn√≥stico de columnas: {nombre_tabla}")
    print("‚Äî" * 60)

    columnas_presentes = df.columns.tolist()
    todo_ok = True

    for col in columnas_esperadas:
        if col in columnas_presentes:
            print(f"‚úîÔ∏è Columna presente: {col}")
        else:
            print(f"‚ùå Falta la columna: {col}")
            todo_ok = False

    if todo_ok:
        print(f"\n‚úÖ {nombre_tabla}: Todas las columnas requeridas est√°n presentes.")
    else:
        print(f"\n‚ö†Ô∏è {nombre_tabla}: Hay columnas faltantes. Revisar el archivo Excel.")

    return todo_ok


columnas_requeridas = {
    "clientes": ["id_cliente", "nombre_cliente", "email", "ciudad", "fecha_alta"],
    "productos": ["id_producto", "nombre_producto", "categoria", "precio_unitario"],
    "ventas": ["id_venta", "fecha", "id_cliente", "nombre_cliente", "email", "medio_pago"],
    "detalle_ventas": ["id_venta", "id_producto", "nombre_producto", "cantidad", "precio_unitario", "importe"]
}

tablas_locales = {
    "clientes": clientes,
    "productos": productos,
    "ventas": ventas,
    "detalle_ventas": detalle
}

for nombre, df in tablas_locales.items():
    diagnostico_columnas(df, columnas_requeridas[nombre], nombre)



# ================================================================================================
# üßº FASE 2 ‚Äî LIMPIEZA Y NORMALIZACI√ìN
# ================================================================================================

def convertir_a_numerico(col):
    """Convierte una columna a tipo num√©rico eliminando errores."""
    return pd.to_numeric(col, errors='coerce')

def limpiar_dataframe(df, id_col):
    """Elimina duplicados y nulos seg√∫n columna id_col."""
    df = df.drop_duplicates(subset=[id_col])
    df = df.dropna(subset=[id_col])
    return df

print("\n==========================================================================================")
print("üßº LIMPIEZA ‚Äî CLIENTES")
print("==========================================================================================")
clientes = limpiar_dataframe(clientes, "id_cliente")
print(f"Clientes v√°lidos: {len(clientes)}")

print("\n==========================================================================================")
print("üßº LIMPIEZA ‚Äî PRODUCTOS")
print("==========================================================================================")
productos = limpiar_dataframe(productos, "id_producto")
print(f"Productos v√°lidos: {len(productos)}")

print("\n==========================================================================================")
print("üßº LIMPIEZA ‚Äî VENTAS")
print("==========================================================================================")
ventas = limpiar_dataframe(ventas, "id_venta")
if "fecha" in ventas.columns:
    ventas["fecha"] = pd.to_datetime(ventas["fecha"], errors="coerce")
print(f"Ventas v√°lidas: {len(ventas)}")

print("\n==========================================================================================")
print("üßº LIMPIEZA ‚Äî DETALLE DE VENTAS")
print("==========================================================================================")
detalle = limpiar_dataframe(detalle, "id_venta")
detalle["precio_unitario"] = convertir_a_numerico(detalle["precio_unitario"])
detalle["importe"] = convertir_a_numerico(detalle["importe"])
print(f"Detalle de ventas v√°lido: {len(detalle)}")

# Exportar tablas limpias
clientes.to_excel(os.path.join(carpeta_salida, "clientes_limpio.xlsx"), index=False)
productos.to_excel(os.path.join(carpeta_salida, "productos_limpio.xlsx"), index=False)
ventas.to_excel(os.path.join(carpeta_salida, "ventas_limpio.xlsx"), index=False)
detalle.to_excel(os.path.join(carpeta_salida, "detalle_ventas_limpio.xlsx"), index=False)

print("\n‚úÖ Tablas limpias exportadas correctamente.")

# ================================================================================================
# üè∑Ô∏è FASE 3 ‚Äî CATEGORIZACI√ìN DEFINITIVA (FUNCI√ìN COMPLETA)
# ================================================================================================

def recategorizar_definitivo(producto):
    p = str(producto).lower().strip()

    # --- Bebidas (no alcoh√≥licas) ---
    if any(x in p for x in ['coca cola','pepsi','sprite','fanta']):
        return ('Alimentos','Bebidas gaseosas','Gaseosas saborizadas')
    if 'energ√©tica' in p or 'nitro' in p:
        return ('Alimentos','Bebidas gaseosas','Bebidas energ√©ticas')
    if 'agua mineral' in p or 'agua con' in p:
        return ('Alimentos','Bebidas sin gas','Aguas minerales')
    if 'jugo de' in p and 'polvo' not in p:
        return ('Alimentos','Bebidas sin gas','Jugos naturales')
    if 'jugo en polvo' in p:
        return ('Alimentos','Bebidas instant√°neas','Jugos en polvo')
    if 'yerba' in p:
        return ('Alimentos','Infusiones','Yerba mate')
    if 'caf√©' in p:
        return ('Alimentos','Infusiones','Caf√©')
    if 't√© ' in p or p.startswith('t√©'):
        return ('Alimentos','Infusiones','T√©')

    # --- Bebidas alcoh√≥licas ---
    if 'cerveza' in p:
        return ('Alimentos','Bebidas alcoh√≥licas','Cervezas')
    if 'vino' in p:
        return ('Alimentos','Bebidas alcoh√≥licas','Vinos')
    if 'sidra' in p:
        return ('Alimentos','Bebidas alcoh√≥licas','Sidras y espumantes')
    if any(x in p for x in ['vodka','ron','gin','whisky']):
        return ('Alimentos','Bebidas alcoh√≥licas','Destilados')
    if 'fernet' in p or 'licor' in p:
        return ('Alimentos','Bebidas alcoh√≥licas','Licores y aperitivos')

    # --- L√°cteos ---
    if 'leche' in p and 'dulce' not in p:
        return ('Alimentos','L√°cteos','Leches')
    if 'yogur' in p or 'yogurt' in p:
        return ('Alimentos','L√°cteos','Yogures')
    if 'queso' in p:
        return ('Alimentos','L√°cteos','Quesos')
    if 'manteca' in p or 'mantequilla' in p:
        return ('Alimentos','L√°cteos','Manteca y margarina')

    # --- Panader√≠a ---
    if 'pan lactal' in p or 'pan de' in p:
        return ('Alimentos','Panader√≠a','Panes')
    if 'medialuna' in p:
        return ('Alimentos','Panader√≠a','Medialunas')
    if 'bizcocho' in p:
        return ('Alimentos','Panader√≠a','Bizcochos')
    if 'galletita' in p or 'galleta' in p:
        return ('Alimentos','Panader√≠a','Galletas')
    if 'alfajor' in p:
        return ('Alimentos','Panader√≠a','Alfajores')

    # --- Snacks y golosinas ---
    if 'papa' in p and 'frita' in p:
        return ('Alimentos','Snacks y golosinas','Papas fritas')
    if 'man√≠' in p or 'frutos secos' in p:
        return ('Alimentos','Snacks y golosinas','Frutos secos')
    if 'chocolate' in p and any(x in p for x in ['amargo','leche','100g']):
        return ('Alimentos','Snacks y golosinas','Chocolates')
    if 'turr√≥n' in p:
        return ('Alimentos','Snacks y golosinas','Turrones')
    if 'barrita' in p:
        return ('Alimentos','Snacks y golosinas','Barritas energ√©ticas')
    if 'caramelo' in p:
        return ('Alimentos','Snacks y golosinas','Caramelos')
    if 'chicle' in p:
        return ('Alimentos','Snacks y golosinas','Chicles')
    if 'chupet√≠n' in p:
        return ('Alimentos','Snacks y golosinas','Chupetines')
    if 'granola' in p:
        return ('Alimentos','Snacks y golosinas','Cereales y granolas')

    # --- Dulces y postres ---
    if 'dulce de leche' in p:
        return ('Alimentos','Dulces y postres','Dulce de leche')
    if 'mermelada' in p:
        return ('Alimentos','Dulces y postres','Mermeladas')
    if 'helado' in p:
        return ('Alimentos','Dulces y postres','Helados')
    if 'miel' in p:
        return ('Alimentos','Dulces y postres','Miel')

    # --- Aceites y condimentos / salsas ---
    if 'aceite' in p:
        return ('Alimentos','Aceites y condimentos','Aceites vegetales')
    if 'vinagre' in p:
        return ('Alimentos','Aceites y condimentos','Vinagres')
    if 'salsa' in p:
        return ('Alimentos','Salsas y aderezos','Salsas b√°sicas')

    # --- Granos/Legumbres/Harinas/Pastas/Avena ---
    if 'arroz' in p:
        return ('Alimentos','Granos y legumbres','Arroces')
    if 'lenteja' in p:
        return ('Alimentos','Granos y legumbres','Lentejas')
    if 'garbanzo' in p:
        return ('Alimentos','Granos y legumbres','Garbanzos')
    if 'poroto' in p or 'frijol' in p:
        return ('Alimentos','Granos y legumbres','Porotos')
    if 'fideo' in p or 'spaghetti' in p or 'pasta' in p:
        return ('Alimentos','Pastas y harinas','Pastas secas')
    if 'harina' in p:
        return ('Alimentos','Pastas y harinas','Harinas')
    if 'avena' in p:
        return ('Alimentos','Cereales y derivados','Avena')

    # --- Despensa b√°sica: az√∫car/sal/stevia ---
    if 'az√∫car' in p or 'azucar' in p:
        return ('Alimentos','Despensa b√°sica','Az√∫car')
    if 'sal fina' in p or 'sal gruesa' in p or p.startswith('sal'):
        return ('Alimentos','Despensa b√°sica','Sal')
    if 'stevia' in p or 'edulcorante' in p:
        return ('Alimentos','Endulzantes y condimentos','Endulzantes naturales')

    # --- Conservas/encurtidos ---
    if 'aceituna' in p:
        return ('Alimentos','Conservas y encurtidos','Aceitunas')

    # --- Sopas y caldos ---
    if 'sopa instant√°nea' in p or 'sopa deshidratada' in p:
        return ('Alimentos','Sopas y caldos','Sopas instant√°neas')
    if 'caldo concentrado' in p or re.search(r'\bcaldo\b', p):
        return ('Alimentos','Sopas y caldos','Caldos concentrados')

    # --- Congelados ---
    if 'pizza' in p and 'congelada' in p:
        return ('Alimentos','Congelados','Pizzas y comidas listas')
    if 'empanada' in p and 'congelada' in p:
        return ('Alimentos','Congelados','Empanadas y snacks listos')
    if 'verdura' in p and 'congelada' in p:
        return ('Alimentos','Congelados','Verduras congeladas')
    if 'hamburguesa' in p and 'congelada' in p:
        return ('Alimentos','Congelados','Carnes procesadas')

    # --- Limpieza (hogar) ---
    if 'detergente' in p:
        return ('Limpieza','Limpieza de ropa','Detergentes')
    if 'lavandina' in p or 'blanqueador' in p:
        return ('Limpieza','Limpieza de ropa','Blanqueadores')
    if 'suavizante' in p:
        return ('Limpieza','Limpieza de ropa','Suavizantes')
    if 'limpiavidrio' in p:
        return ('Limpieza','Limpieza del hogar','Limpieza de superficies')
    if 'desengrasante' in p:
        return ('Limpieza','Limpieza del hogar','Desengrasantes')
    if 'esponja' in p:
        return ('Limpieza','Accesorios de limpieza','Esponjas')
    if 'trapo' in p:
        return ('Limpieza','Accesorios de limpieza','Trapos y pa√±os')

    # --- Cuidado personal ---
    if 'jab√≥n de tocador' in p or 'jab√≥n en barra' in p:
        return ('Cuidado personal','Higiene corporal','Jabones')
    if 'shampoo' in p or 'champ√∫' in p:
        return ('Cuidado personal','Cuidado del cabello','Shampoo')
    if 'mascarilla capilar' in p or 'tratamiento capilar' in p:
        return ('Cuidado personal','Cuidado del cabello','Mascarillas capilares')
    if 'toallas h√∫medas x50' in p or 'toallita h√∫meda' in p:
        return ('Cuidado personal','Higiene corporal','Toallas h√∫medas')
    if 'desodorante' in p:
        return ('Cuidado personal','Higiene corporal','Desodorantes')
    if 'crema dental' in p or 'pasta dental' in p:
        return ('Cuidado personal','Higiene bucal','Cremas dentales')
    if 'cepillo de dientes' in p or 'cepillo dental' in p:
        return ('Cuidado personal','Higiene bucal','Cepillos dentales')
    if 'hilo dental' in p:
        return ('Cuidado personal','Higiene bucal','Hilos dentales')

    # --- Papeler√≠a y desechables ---
    if 'papel higi√©nico' in p:
        return ('Papeler√≠a y desechables','Papel higi√©nico','Papel higi√©nico')
    if 'servilleta' in p:
        return ('Papeler√≠a y desechables','Servilletas','Servilletas de papel')

    # Default
    return ('Sin categorizar', 'Revisar', 'Producto no identificado')

# Aplicar categorizaci√≥n
productos[["categoria_principal","categoria_secundaria","categoria_terciaria"]] = (
    productos["nombre_producto"].apply(lambda x: pd.Series(recategorizar_definitivo(x)))
)

productos.to_excel(os.path.join(carpeta_salida, "productos_categorizados_final.xlsx"), index=False)
print("\n‚úÖ Categorizaci√≥n aplicada correctamente a los 100 productos.")

# ================================================================================================
# üìä FASE 4 ‚Äî AN√ÅLISIS ESTAD√çSTICO DESCRIPTIVO
# ================================================================================================

print("\nüìà C√°lculo de estad√≠sticas b√°sicas...")
stats = productos.describe(include="all").transpose()
display(stats)

print("\nüìä An√°lisis de distribuci√≥n de precios...")
plt.figure(figsize=(8,5))
sns.histplot(productos["precio_unitario"], kde=True, bins=15)
plt.title("Distribuci√≥n de precios unitarios")
plt.xlabel("Precio unitario")
plt.ylabel("Frecuencia")
plt.show()

print("\nüìä Boxplot de precios por categor√≠a principal...")
plt.figure(figsize=(10,5))
sns.boxplot(x="categoria_principal", y="precio_unitario", data=productos)
plt.title("Distribuci√≥n de precios por categor√≠a principal")
plt.xticks(rotation=45)
plt.show()

print("\nüìä Dispersi√≥n Precio vs Producto ID...")
plt.figure(figsize=(8,5))
plt.scatter(productos["id_producto"], productos["precio_unitario"], alpha=0.7)
plt.title("Dispersi√≥n entre ID de producto y Precio unitario")
plt.xlabel("ID Producto")
plt.ylabel("Precio Unitario")
plt.show()

# Guardar an√°lisis
stats.to_excel(os.path.join(carpeta_salida, "analisis_estadistico.xlsx"))

print("\n‚úÖ Fase 4 completada ‚Äî an√°lisis y gr√°ficos generados correctamente.")

"""# DESDE AQUI"""

# --- Funciones utilitarias y de limpieza --- #

def normalizar_columnas(df):
    """Normaliza nombres de columnas: min√∫sculas, sin espacios ni tildes."""
    df.columns = (
        df.columns.str.strip()
        .str.lower()
        .str.normalize('NFKD')
        .str.encode('ascii', errors='ignore')
        .str.decode('utf-8')
        .str.replace(r'[^0-9a-zA-Z_]', '_', regex=True)
    )
    return df


def corregir_tipos(df, estructura):
    """Corrige tipos de columnas seg√∫n estructura dada."""
    conv = {
        "int": lambda s: pd.to_numeric(s, errors="coerce").astype("Int64"),
        "float": lambda s: pd.to_numeric(s, errors="coerce"),
        "str": lambda s: s.astype("string").str.strip(),
        "date": lambda s: pd.to_datetime(s, errors="coerce")
    }
    for col, tipo in estructura.items():
        if col in df.columns:
            df[col] = conv.get(tipo, lambda s: s)(df[col])
    return df


def semaforo(valor, bajo=5, alto=20):
    """Devuelve emoji tipo sem√°foro seg√∫n valor."""
    return "üü¢" if valor == 0 else "üü°" if valor <= bajo else "üî¥"


# --- Reporte de calidad --- #

def reporte_calidad(df, nombre):
    """Genera resumen y detalle de calidad de una tabla."""
    nulos = df.isnull().sum().sum()
    porc_nulos = round(nulos / df.size * 100, 2)
    duplicados = df.duplicated().sum()

    resumen = {
        "Tabla": nombre,
        "Filas": len(df),
        "Columnas": df.shape[1],
        "Nulos Totales": nulos,
        "No Nulos Totales": df.notnull().sum().sum(),
        "Duplicados": duplicados,
        "Calidad Nulos": semaforo(porc_nulos),
        "Calidad Duplicados": semaforo(duplicados)
    }

    detalle = pd.DataFrame({
        "Tabla": nombre,
        "Columna": df.columns,
        "Tipo": df.dtypes.values,
        "Nulos": df.isnull().sum().values,
        "No Nulos": df.notnull().sum().values,
        "Total Filas": len(df),
        "Sem√°foro Nulos": [semaforo(v) for v in df.isnull().sum().values]
    })
    return resumen, detalle


# --- Definici√≥n de estructuras --- #

estructuras = {
    "productos": {
        "id_producto": "int", "nombre_producto": "str",
        "categoria": "str", "precio_unitario": "float"
    },
    "clientes": {
        "id_cliente": "int", "nombre_cliente": "str",
        "email": "str", "ciudad": "str", "fecha_alta": "date"
    },
    "ventas": {
        "id_venta": "int", "fecha": "date", "id_cliente": "int",
        "nombre_cliente": "str", "email": "str", "medio_pago": "str"
    },
    "detalle_ventas": {
        "id_venta": "int", "id_producto": "int",
        "nombre_producto": "str", "cantidad": "int",
        "precio_unitario": "float", "importe": "float"
    }
}


tablas = {
    "productos": productos,
    "clientes": clientes,
    "ventas": ventas,
    "detalle_ventas": detalle
}


# --- Limpieza y generaci√≥n de reportes --- #

resumenes, detalles = [], []

for nombre, df in tablas.items():
    df = normalizar_columnas(df)
    df = corregir_tipos(df, estructuras[nombre])
    df = df.drop_duplicates(ignore_index=True)

    tablas[nombre] = df
    r, d = reporte_calidad(df, nombre)
    resumenes.append(r)
    detalles.append(d)

    # --- Mostrar resultados parciales ---
reporte_resumen = pd.DataFrame(resumenes)
reporte_detalle = pd.concat(detalles, ignore_index=True)

print("=== RESUMEN GENERAL POR TABLA ===")
display(reporte_resumen)

print("\n=== DETALLE POR COLUMNA ===")
display(reporte_detalle)

# ARCHIVO CATEGORIAS REF
# Si lo ejecuto una segunda vez, se ordena alfabeticamente de a-z
# Podemos generar autom√°ticamente un archivo de referencia inicial, con todos los productos actuales.
# Exportar tu tabla actual de productos para crear la referencia inicial

if "productos" in tablas:
    df_productos = tablas["productos"]

    # Verificar que existan las columnas necesarias
    if {"nombre_producto", "categoria"}.issubset(df_productos.columns):

        # Crear DataFrame base con jerarqu√≠a de categor√≠as
        df_ref_inicial = df_productos[["nombre_producto", "categoria"]].rename(
            columns={"categoria": "categoria_n_1"}
        )

        # Agregar columnas vac√≠as para niveles 2 y 3
        df_ref_inicial["categoria_n_2"] = ""
        df_ref_inicial["categoria_n_3"] = ""

        # Ruta del archivo de referencia
        salida_ref = os.path.join(ruta_base, "categorias_ref.xlsx")

        # --- L√≥gica principal: conservar ediciones si el archivo ya existe ---
        if os.path.exists(salida_ref):
            print(f"‚ÑπÔ∏è Archivo existente detectado: {salida_ref}")
            print("   ‚ûú Se mantendr√°n las ediciones manuales y se agregar√°n productos nuevos...")

            # Leer archivo existente (con ediciones previas)
            df_existente = pd.read_excel(salida_ref)

            # Combinar ambos tablas
            df_final = pd.merge(
                df_existente,
                df_ref_inicial,
                on="nombre_producto",
                how="outer",
                suffixes=("", "_nuevo")
            )

            # Si hay categor√≠a nueva, pero la existente ya est√° editada, se conserva la editada
            for col in ["categoria_n_1", "categoria_n_2", "categoria_n_3"]:
                col_nuevo = f"{col}_nuevo"
                if col_nuevo in df_final.columns:
                    df_final[col] = df_final[col].fillna(df_final[col_nuevo])

            # üîç Detectar productos nuevos sin categor√≠a asignada
            nuevos = df_final[df_final["categoria_n_1"].isna()]
            if not nuevos.empty:
                print(f"üÜï {len(nuevos)} productos nuevos agregados sin categor√≠a asignada.")
                print("   ‚ûú Revisa 'categorias_ref.xlsx' para completarlas manualmente.")

            # Limpiar columnas auxiliares (_nuevo)
            df_final = df_final[[c for c in df_final.columns if not c.endswith("_nuevo")]]

            # Guardar archivo actualizado
            df_final.to_excel(salida_ref, index=False)
            print(f"‚úÖ Archivo actualizado (manteniendo ediciones) en:\n   {salida_ref}")

        else:
            # Si no existe, se crea desde cero
            df_ref_inicial.to_excel(salida_ref, index=False)
            print(f"üìÅ Archivo nuevo 'categorias_ref.xlsx' creado con todas las categor√≠as actuales en:\n   {salida_ref}")

    else:
        print("‚ö†Ô∏è 'productos' no tiene las columnas 'nombre_producto' y 'categoria'.")
else:
    print("‚ö†Ô∏è No se encontr√≥ el archivo 'productos' en el diccionario de tablas.")

# Actualizaci√≥n de categor√≠as jer√°rquicas desde archivo de referencia
ruta_ref = os.path.join(ruta_base, "categorias_ref.xlsx")

if "productos" in tablas:
    df_productos = tablas["productos"].copy()

    # Eliminar columna vieja si existe
    if "categoria" in df_productos.columns:
        df_productos.drop(columns=["categoria"], inplace=True)

    # Asegurar que las columnas jer√°rquicas existan
    for col in ["categoria_n_1", "categoria_n_2", "categoria_n_3"]:
        if col not in df_productos.columns:
            df_productos[col] = pd.NA

    if os.path.exists(ruta_ref):
        df_ref = pd.read_excel(ruta_ref)

        # === Verificaci√≥n de columnas necesarias ===
        columnas_necesarias = {"nombre_producto", "categoria_n_1", "categoria_n_2", "categoria_n_3"}
        if not columnas_necesarias.issubset(df_ref.columns):
            raise ValueError(f"‚ö†Ô∏è El archivo categorias_ref.xlsx debe tener las columnas: {', '.join(columnas_necesarias)}")

        # === Normalizar nombres de productos para mejorar coincidencias ===
        def normalizar_texto(s):
            if pd.isna(s):
                return ""
            s = str(s).lower()
            s = unicodedata.normalize("NFKD", s).encode("ASCII", "ignore").decode("utf-8")
            s = re.sub(r"[^a-z0-9 ]", " ", s)
            s = re.sub(r"\s+", " ", s).strip()
            return s

        df_productos["_nombre_norm"] = df_productos["nombre_producto"].apply(normalizar_texto)
        df_ref["_nombre_norm"] = df_ref["nombre_producto"].apply(normalizar_texto)

        # === Combinar por nombre normalizado ===
        df_merged = df_productos.merge(
            df_ref[["_nombre_norm", "categoria_n_1", "categoria_n_2", "categoria_n_3"]],
            on="_nombre_norm",
            how="left",
            suffixes=("", "_ref")
        )

        # === Reemplazar categor√≠as solo donde existan valores en el archivo de referencia ===
        for col in ["categoria_n_1", "categoria_n_2", "categoria_n_3"]:
            col_ref = f"{col}_ref"
            if col_ref in df_merged.columns:
                df_merged[col] = df_merged[col].fillna(df_merged[col_ref])

        # === Limpiar columnas auxiliares ===
        df_merged.drop(columns=[c for c in df_merged.columns if c.endswith("_ref") or c == "_nombre_norm"], inplace=True, errors="ignore")

        # === Actualizar dentro del diccionario ===
        tablas["productos"] = df_merged
        df_productos = df_merged

        # === Guardar archivo actualizado ===
        salida = os.path.join(ruta_base, "productos_actualizado.xlsx")
        df_productos.to_excel(salida, index=False)

        print("‚úÖ Categor√≠as jer√°rquicas actualizadas correctamente con normalizaci√≥n.")
        print(f"üìÅ Archivo guardado en: {salida}")

        # === Mostrar resumen ===
        total = len(df_productos)
        actualizados = sum(df_productos[col].notna().sum() for col in ["categoria_n_1", "categoria_n_2", "categoria_n_3"])
        print(f"Total productos: {total} | Categor√≠as actualizadas (en cualquier nivel): {actualizados}")

    else:
        print("‚ö†Ô∏è No se encontr√≥ el archivo categorias_ref.xlsx. Ejec√∫talo primero para generarlo.")
else:
    print("‚ö†Ô∏è No se encontr√≥ el archivo 'productos' en el diccionario de tablas.")

# ================================================================================================
# üìä FASE 5 ‚Äî VISUALIZACI√ìN AVANZADA Y DASHBOARD DESCRIPTIVO
# ================================================================================================

print("\nüìä Iniciando visualizaci√≥n avanzada...")

productos = tablas["productos"].copy()

# Crear carpeta de salida para gr√°ficos
carpeta_graficos = os.path.join(ruta_base, "graficos")
os.makedirs(carpeta_graficos, exist_ok=True)

# -----------------------------------------------------------------------------
# üìà 1. Promedio de precios por categor√≠a principal
# -----------------------------------------------------------------------------
print("\nüìä Generando gr√°fico: promedio de precios por categor√≠a principal...")

if "categoria_n_1" in productos.columns:
    resumen_cat = (
        productos.groupby("categoria_n_1", dropna=False)["precio_unitario"]
        .agg(["count", "mean", "std", "min", "max"])
        .reset_index()
        .sort_values(by="mean", ascending=False)
    )

    plt.figure(figsize=(10,6))
    plt.bar(resumen_cat["categoria_n_1"], resumen_cat["mean"])
    plt.title("Promedio de precios unitarios por categor√≠a principal")
    plt.xlabel("Categor√≠a principal (categoria_n_1)")
    plt.ylabel("Precio promedio")
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()

    # Guardar gr√°fico y tabla resumen
    resumen_cat.to_excel(os.path.join(carpeta_graficos, "resumen_precios_categoria.xlsx"), index=False)
    plt.savefig(os.path.join(carpeta_graficos, "promedio_precios_categoria.png"), dpi=300)
    print("‚úÖ Promedio de precios por categor√≠a generado y guardado.")
else:
    print("‚ö†Ô∏è No se encontr√≥ la columna 'categoria_n_1'.")

# -----------------------------------------------------------------------------
# üìä 2. Distribuci√≥n de cantidad de productos por categor√≠a principal
# -----------------------------------------------------------------------------
print("\nüìä Generando gr√°fico: cantidad de productos por categor√≠a...")

if "categoria_n_1" in productos.columns:
    conteo_cat = productos["categoria_n_1"].value_counts(dropna=False).reset_index()
    conteo_cat.columns = ["categoria_n_1", "cantidad_productos"]

    plt.figure(figsize=(10,6))
    plt.barh(conteo_cat["categoria_n_1"], conteo_cat["cantidad_productos"])
    plt.title("Cantidad de productos por categor√≠a principal")
    plt.xlabel("Cantidad de productos")
    plt.ylabel("Categor√≠a principal")
    plt.grid(axis="x", linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()

    conteo_cat.to_excel(os.path.join(carpeta_graficos, "cantidad_productos_categoria.xlsx"), index=False)
    plt.savefig(os.path.join(carpeta_graficos, "cantidad_productos_categoria.png"), dpi=300)
    print("‚úÖ Conteo de productos por categor√≠a generado y guardado.")
else:
    print("‚ö†Ô∏è No se encontr√≥ la columna 'categoria_n_1'.")

# -----------------------------------------------------------------------------
# üìä 3. Top 10 productos m√°s caros
# -----------------------------------------------------------------------------
print("\nüìä Generando gr√°fico: Top 10 productos m√°s caros...")

top10 = productos.nlargest(10, "precio_unitario")[["nombre_producto", "precio_unitario", "categoria_n_1"]]

plt.figure(figsize=(10,6))
plt.barh(top10["nombre_producto"], top10["precio_unitario"], color="skyblue")
plt.title("Top 10 productos m√°s caros")
plt.xlabel("Precio unitario")
plt.ylabel("Producto")
plt.gca().invert_yaxis()  # Muestra el m√°s caro arriba
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

top10.to_excel(os.path.join(carpeta_graficos, "top10_productos_caros.xlsx"), index=False)
plt.savefig(os.path.join(carpeta_graficos, "top10_productos_caros.png"), dpi=300)
print("‚úÖ Gr√°fico Top 10 productos m√°s caros generado y guardado.")

# -----------------------------------------------------------------------------
# üìä 4. Correlaci√≥n num√©rica entre variables
# -----------------------------------------------------------------------------
print("\nüìä Calculando matriz de correlaci√≥n...")

num_cols = productos.select_dtypes(include=["number"]).columns
if len(num_cols) > 1:
    corr = productos[num_cols].corr()

    plt.figure(figsize=(6,5))
    plt.imshow(corr, cmap="coolwarm", interpolation="none")
    plt.colorbar(label="Correlaci√≥n")
    plt.xticks(range(len(num_cols)), num_cols, rotation=45, ha="right")
    plt.yticks(range(len(num_cols)), num_cols)
    plt.title("Matriz de correlaci√≥n num√©rica")
    plt.tight_layout()
    plt.show()

    corr.to_excel(os.path.join(carpeta_graficos, "matriz_correlacion.xlsx"))
    plt.savefig(os.path.join(carpeta_graficos, "matriz_correlacion.png"), dpi=300)
    print("‚úÖ Matriz de correlaci√≥n generada y guardada.")
else:
    print("‚ö†Ô∏è No hay suficientes columnas num√©ricas para correlaci√≥n.")

# -----------------------------------------------------------------------------
# ‚úÖ Finalizaci√≥n
# -----------------------------------------------------------------------------
print("\nüéØ Fase 5 completada ‚Äî gr√°ficos comparativos y dashboard descriptivo generados correctamente.")
print(f"üìÇ Todos los resultados se guardaron en: {carpeta_graficos}")
```

## Fase 6: Machine Learning ‚Äî Predicci√≥n del Importe de una Venta

En esta FASE del proyecto, aplicamos t√©cnicas de **Machine Learning** para construir un modelo de **Regresi√≥n Supervisada** cuyo objetivo principal es **predecir el importe total de una venta** en la Tienda Aurelion.

Para preparar la informaci√≥n, unificamos todas las tablas del proyecto (productos, clientes, ventas y detalle de ventas) y generamos variables nuevas y relevantes como el a√±o, el mes o la antig√ºedad del cliente.

El modelo seleccionado es la **Regresi√≥n Lineal**, un algoritmo supervisado ideal para predecir valores num√©ricos. Este modelo permite:

* **Estimar ingresos futuros** y analizar cu√°nto podr√≠a costar una venta bajo distintas condiciones.
* **Detectar patrones de compra** y analizar c√≥mo influyen variables como el precio unitario, la cantidad o la categor√≠a del producto.
* **Analizar qu√© caracter√≠sticas influyen m√°s** en el importe total de la venta. **texto en negrita**
 * **Evaluar estrategias de precios** y mejorar la comprensi√≥n del comportamiento de las ventas.


Finalmente, el **dashboard** de esta fase es clave, ya que muestra el rendimiento del modelo a trav√©s de m√©tricas como MAE, RMSE y $R^2$. Tambi√©n incluye la comparaci√≥n entre los valores reales y los predichos, la distribuci√≥n de errores y, lo m√°s importante, las variables que resultaron ser m√°s influyentes en la determinaci√≥n del importe final de la venta.



```python
# ================================================================================================
# FASE 6 ‚Äî MACHINE LEARNING: Predicci√≥n del importe de una venta
# Aporte de dise√±o y c√≥digo: Anah√≠ Gonz√°lez
# ================================================================================================

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("\n Iniciando Fase 6 ‚Äî Machine Learning (Regresi√≥n Lineal sobre importe)...")

# ------------------------------------------------------------------------------
# 1. Recuperaci√≥n segura de tablas desde el diccionario 'tablas'
# ------------------------------------------------------------------------------

df_productos     = tablas["productos"].copy()
df_clientes      = tablas["clientes"].copy()
df_ventas        = tablas["ventas"].copy()
df_detalle       = tablas["detalle_ventas"].copy()

print("Tablas cargadas correctamente desde el diccionario 'tablas'.")

# ------------------------------------------------------------------------------
# 2. Construcci√≥n del dataset unificado
# ------------------------------------------------------------------------------

print("\nUnificando tablas necesarias para Machine Learning...")

df_ml = df_detalle.merge(
    df_ventas[["id_venta", "id_cliente", "fecha"]],
    on="id_venta",
    how="left"
)

# columnas desde productos
cols_prod = ["id_producto", "precio_unitario"]
for col_cat in ["categoria_principal", "categoria_secundaria", "categoria_n_1", "categoria_n_2", "categoria_n_3"]:
    if col_cat in df_productos.columns and col_cat not in cols_prod:
        cols_prod.append(col_cat)

df_ml = df_ml.merge(
    df_productos[cols_prod],
    on="id_producto",
    how="left"
)

# columnas desde clientes
cols_cli = ["id_cliente"]
for col in ["ciudad", "fecha_alta"]:
    if col in df_clientes.columns and col not in cols_cli:
        cols_cli.append(col)

df_ml = df_ml.merge(
    df_clientes[cols_cli],
    on="id_cliente",
    how="left"
)

print(f"Dataset final para ML: {df_ml.shape[0]} filas, {df_ml.shape[1]} columnas.")


# ------------------------------------------------------------------------------
# 3. Ingenier√≠a de caracter√≠sticas
# ------------------------------------------------------------------------------

print("\nCreando nuevas variables (features)...")

if "fecha" in df_ml.columns:
    df_ml["anio_venta"] = df_ml["fecha"].dt.year
    df_ml["mes_venta"] = df_ml["fecha"].dt.month

if "fecha" in df_ml.columns and "fecha_alta" in df_ml.columns:
    df_ml["antiguedad_cliente_dias"] = (df_ml["fecha"] - df_ml["fecha_alta"]).dt.days


# ------------------------------------------------------------------------------
# 4. Definici√≥n de variables X e y
# ------------------------------------------------------------------------------

print("\nDefiniendo X (entradas) e y (salida)...")

y = df_ml["importe"]

features_numericas = [
    "precio_unitario",
    "cantidad",
    "anio_venta",
    "mes_venta",
    "antiguedad_cliente_dias"
]
# Filter for actual columns present in df_ml after merges and feature engineering
features_numericas = [f for f in features_numericas if f in df_ml.columns]

features_categoricas = []
for col_cat in ["categoria_principal", "categoria_secundaria", "categoria_n_1", "categoria_n_2", "categoria_n_3", "ciudad"]:
    if col_cat in df_ml.columns:
        features_categoricas.append(col_cat)

# Select features from df_ml before handling NaNs
X = df_ml[features_numericas + features_categoricas].copy()

# --- Diagn√≥stico de NaNs antes del filtrado ---
print("\nüîé Valores nulos por columna antes de procesar:\n", X.isnull().sum())

# ------------------------------------------------------------------------------
# 5. Manejo de valores nulos
# ------------------------------------------------------------------------------

print("\nManejando valores nulos...")

# Drop rows where the target variable 'importe' is null
initial_rows = len(df_ml)
df_ml.dropna(subset=["importe"], inplace=True)
if len(df_ml) < initial_rows:
    print(f"   ‚ûú Se eliminaron {initial_rows - len(df_ml)} filas con 'importe' nulo.")

# Fill NaNs for categorical features with a placeholder
for cat_col in features_categoricas:
    if cat_col in X.columns:
        X[cat_col] = X[cat_col].fillna('DESCONOCIDO') # Changed to explicit assignment

# Drop rows where numerical features still have NaNs
X.dropna(subset=features_numericas, inplace=True)

# Ensure y matches X after dropping NaNs
# Re-align y to X's index after dropping NaNs in X
y = y.loc[X.index]

print(f"Datos disponibles para entrenamiento despu√©s de manejo de nulos: {X.shape[0]} filas.")

# ------------------------------------------------------------------------------
# 6. Codificaci√≥n de variables categ√≥ricas
# ------------------------------------------------------------------------------

print("\nAplicando One-Hot Encoding...")

X_encoded = pd.get_dummies(X, columns=features_categoricas, drop_first=True)

print(f"Matriz final de caracter√≠sticas: {X_encoded.shape[1]} columnas.")


# ------------------------------------------------------------------------------
# 7. Divisi√≥n Train/Test
# ------------------------------------------------------------------------------

print("\nDividiendo en entrenamiento (80%) y prueba (20%)...")

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print(f"Entrenamiento: {X_train.shape[0]} filas ‚Äî Prueba: {X_test.shape[0]} filas.")


# ------------------------------------------------------------------------------
# 8. Entrenamiento del modelo
# ------------------------------------------------------------------------------

print("\nEntrenando modelo de Regresi√≥n Lineal...")

modelo = LinearRegression()
modelo.fit(X_train, y_train)

print("Modelo entrenado correctamente.")


# ------------------------------------------------------------------------------
# 9. Evaluaci√≥n del modelo
# ------------------------------------------------------------------------------

print("\nEvaluando el modelo...")

y_pred = modelo.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2   = r2_score(y_test, y_pred)

print("\nResultados del modelo:")
print(f"MAE :  {mae:.2f}")
print(f"RMSE:  {rmse:.2f}")
print(f"R¬≤  :  {r2:.3f}")


# ------------------------------------------------------------------------------
# 10. Interpretaci√≥n r√°pida
# ------------------------------------------------------------------------------

print("\nInterpretaci√≥n:")
print("- MAE: error promedio en unidades monetarias.")
print("- RMSE: penaliza errores grandes.")
print("- R¬≤: proporci√≥n explicada de la variabilidad del importe.")

print("\nFase 6 completada.")
```

## Introducci√≥n a la Fase 7 ‚Äî Predicci√≥n de ventas diarias

Adem√°s del modelo que predice el importe por venta, armamos un segundo modelo orientado a estimar **cu√°ntas ventas tendremos por d√≠a**.  

El dashboard correspondiente muestra:  
- las m√©tricas del modelo,  
- la comparaci√≥n entre ventas reales y predichas por d√≠a,  
- la distribuci√≥n de los errores,  
- y las variables que influyen m√°s en las ventas diarias, como el d√≠a de la semana o la cantidad de clientes activos.

Este modelo es √∫til para **planificar stock, organizar el personal, anticipar d√≠as de mayor movimiento** y detectar patrones semanales o estacionales.


```python
# ================================================================================================
# FASE 7 ‚Äî MACHINE LEARNING: Predicci√≥n de ventas diarias (Regresi√≥n)
# Modelo alternativo: cantidad de ventas por d√≠a
# ================================================================================================

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("\nIniciando Fase 7 ‚Äî Predicci√≥n de ventas diarias...")

# ------------------------------------------------------------------------------
# 1. Partimos de la tabla de ventas
# ------------------------------------------------------------------------------

df_ventas = tablas["ventas"].copy()

# Aseguramos que 'fecha' sea tipo datetime
df_ventas["fecha"] = pd.to_datetime(df_ventas["fecha"])

# ------------------------------------------------------------------------------
# 2. Agregaci√≥n diaria: cu√°ntas ventas y cu√°ntos clientes hubo por d√≠a
# ------------------------------------------------------------------------------

df_diario = (
    df_ventas
    .groupby("fecha")
    .agg(
        ventas_totales=("id_venta", "nunique"),      # cantidad de ventas (comprobantes) por d√≠a
        clientes_activos=("id_cliente", "nunique")   # cantidad de clientes distintos por d√≠a
    )
    .reset_index()
)

print("\nVista de la tabla diaria (primeras filas):")
display(df_diario.head())

# ------------------------------------------------------------------------------
# 3. Ingenier√≠a de variables de fecha
# ------------------------------------------------------------------------------

df_diario["anio"] = df_diario["fecha"].dt.year
df_diario["mes"] = df_diario["fecha"].dt.month
df_diario["dia"] = df_diario["fecha"].dt.day
df_diario["dia_semana"] = df_diario["fecha"].dt.weekday  # 0=Lunes, 6=Domingo
df_diario["es_fin_de_semana"] = df_diario["dia_semana"].isin([5, 6]).astype(int)

# Columna de feriados (de momento todo 0, luego se puede editar a mano)
df_diario["feriado"] = 0

print("\nTabla diaria con variables de fecha:")
display(df_diario.head())

# ------------------------------------------------------------------------------
# 4. Definimos X (entradas) e y (salida)
# ------------------------------------------------------------------------------

# y: cantidad de ventas diarias
y_ventas = df_diario["ventas_totales"]

# X: variables explicativas
features_diarias = [
    "anio",
    "mes",
    "dia_semana",
    "es_fin_de_semana",
    "clientes_activos",
    "feriado"
]

X_ventas = df_diario[features_diarias].copy()

print("\nVariables de entrada (X):", features_diarias)
print("Variable objetivo (y): ventas_totales")

# ------------------------------------------------------------------------------
# 5. Divisi√≥n Train/Test
# ------------------------------------------------------------------------------

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(
    X_ventas,
    y_ventas,
    test_size=0.2,
    random_state=42
)

print(f"\nTrain diario: {X_train_d.shape[0]} filas | Test diario: {X_test_d.shape[0]} filas")

# ------------------------------------------------------------------------------
# 6. Entrenamiento del modelo de regresi√≥n diaria
# ------------------------------------------------------------------------------

modelo_diario = LinearRegression()
modelo_diario.fit(X_train_d, y_train_d)

print("\nModelo de regresi√≥n diaria entrenado correctamente.")

# ------------------------------------------------------------------------------
# 7. Evaluaci√≥n del modelo diario
# ------------------------------------------------------------------------------

y_pred_d = modelo_diario.predict(X_test_d)

mae_d  = mean_absolute_error(y_test_d, y_pred_d)
rmse_d = np.sqrt(mean_squared_error(y_test_d, y_pred_d))
r2_d   = r2_score(y_test_d, y_pred_d)

print("\nResultados del modelo de ventas diarias:")
print(f"MAE  diario : {mae_d:.2f}")
print(f"RMSE diario : {rmse_d:.2f}")
print(f"R¬≤   diario : {r2_d:.3f}")

print("\nInterpretaci√≥n r√°pida para el modelo diario:")
print("- MAE diario: error promedio en cantidad de ventas por d√≠a.")
print("- RMSE diario: penaliza d√≠as donde el modelo se equivoc√≥ mucho.")
print("- R¬≤ diario: proporci√≥n de la variabilidad de las ventas diarias que explica el modelo.")

print("\nFase 7 completada: modelo de predicci√≥n de ventas diarias entrenado y evaluado.")
```

## Fase 8 ‚Äî Clasificaci√≥n de clientes: ¬øvolver√° a comprar o no?

En esta fase incorporamos un modelo de **clasificaci√≥n binaria** para identificar si un cliente volver√° a comprar en la tienda (activo = 1) o si es probable que no regrese (inactivo = 0).  
Este an√°lisis surge de la necesidad de entender por qu√© algunos clientes han dejado de comprar y c√≥mo esto afecta las ventas.

El modelo se basa en datos del historial del cliente, como la cantidad de compras realizadas, el monto promedio, la categor√≠a que m√°s consume y el tiempo transcurrido desde su √∫ltima compra.

El algoritmo elegido fue **Regresi√≥n Log√≠stica**, ideal para este tipo de clasificaci√≥n y para interpretar la importancia de cada variable.  
El rendimiento del modelo se evalu√≥ mediante **accuracy, precision, recall y F1-score**, m√©tricas fundamentales para detectar correctamente a los clientes en riesgo de abandono.

Este modelo permite al negocio anticiparse y tomar decisiones focalizadas para recuperar clientes y mejorar la retenci√≥n.
  





```python
# ================================================================================================
# FASE 8 ‚Äî CLASIFICACI√ìN DE CLIENTES SEG√öN SU TASA DE ABANDONO
# ¬øVolver√° a comprar o no?
# 1 = cliente ACTIVO (sigue comprando)
# 0 = cliente INACTIVO (alta probabilidad de abandono)
# Algoritmo: Regresi√≥n Log√≠stica
# ================================================================================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import matplotlib.pyplot as plt
import seaborn as sns

print("\nIniciando Fase 8 ‚Äî Clasificaci√≥n de clientes seg√∫n tasa de abandono...")

# ------------------------------------------------------------------------------
# 1. Recuperar tablas desde el diccionario 'tablas'
# ------------------------------------------------------------------------------

df_clientes  = tablas["clientes"].copy()
df_ventas    = tablas["ventas"].copy()
df_detalle   = tablas["detalle_ventas"].copy()
df_productos = tablas["productos"].copy()

# Asegurar tipos de fecha
df_ventas["fecha"]      = pd.to_datetime(df_ventas["fecha"])
df_clientes["fecha_alta"] = pd.to_datetime(df_clientes["fecha_alta"])

# ------------------------------------------------------------------------------
# 2. Construcci√≥n de m√©tricas hist√≥ricas por cliente
# ------------------------------------------------------------------------------

# 2.1 Cantidad de compras y fecha de √∫ltima compra
aggr_cliente = (
    df_ventas
    .groupby("id_cliente")
    .agg(
        cant_compras=("id_venta", "nunique"),
        fecha_ultima_compra=("fecha", "max")
    )
    .reset_index()
)

# 2.2 Importe promedio por cliente
df_importe_venta = (
    df_detalle
    .groupby("id_venta")["importe"]
    .sum()
    .reset_index()
)

df_ventas_importe = df_ventas.merge(df_importe_venta, on="id_venta", how="left")

aggr_importe = (
    df_ventas_importe
    .groupby("id_cliente")["importe"]
    .mean()
    .reset_index()
    .rename(columns={"importe": "monto_promedio"})
)

aggr_cliente = aggr_cliente.merge(aggr_importe, on="id_cliente", how="left")

# 2.3 Categor√≠a m√°s comprada por cliente (si existe categoria_principal)
if "categoria_principal" in df_productos.columns:
    df_det_prod = (
        df_detalle
        .merge(
            df_productos[["id_producto", "categoria_principal"]],
            on="id_producto",
            how="left"
        )
        .merge(
            df_ventas[["id_venta", "id_cliente"]],
            on="id_venta",
            how="left"
        )
    )

    conteo_cat = (
        df_det_prod
        .groupby(["id_cliente", "categoria_principal"])
        .size()
        .reset_index(name="n")
    )

    idx_max_cat = conteo_cat.groupby("id_cliente")["n"].idxmax()
    fav_cat = conteo_cat.loc[idx_max_cat, ["id_cliente", "categoria_principal"]].rename(
        columns={"categoria_principal": "categoria_mas_comprada"}
    )

    aggr_cliente = aggr_cliente.merge(fav_cat, on="id_cliente", how="left")
else:
    aggr_cliente["categoria_mas_comprada"] = np.nan

# 2.4 Medio de pago m√°s usado (si existe columna 'medio_pago' en ventas)
if "medio_pago" in df_ventas.columns:
    df_medios = (
        df_ventas
        .groupby(["id_cliente", "medio_pago"])
        .size()
        .reset_index(name="n")
    )
    idx_max_mp = df_medios.groupby("id_cliente")["n"].idxmax()
    fav_mp = df_medios.loc[idx_max_mp, ["id_cliente", "medio_pago"]].rename(
        columns={"medio_pago": "medio_pago_mas_usado"}
    )
    aggr_cliente = aggr_cliente.merge(fav_mp, on="id_cliente", how="left")
else:
    aggr_cliente["medio_pago_mas_usado"] = np.nan

# 2.5 D√≠as desde la √∫ltima compra
max_fecha = df_ventas["fecha"].max()
aggr_cliente["dias_desde_ultima_compra"] = (
    max_fecha - aggr_cliente["fecha_ultima_compra"]
).dt.days

# ------------------------------------------------------------------------------
# 3. Unir con datos de clientes (ciudad, fecha_alta)
# ------------------------------------------------------------------------------

df_tasa_abandono = aggr_cliente.merge(
    df_clientes[["id_cliente", "ciudad", "fecha_alta"]],
    on="id_cliente",
    how="left"
)

# ------------------------------------------------------------------------------
# 4. Crear etiqueta y (tasa de abandono)
# ------------------------------------------------------------------------------
# Objetivo del modelo:
# Detectar clientes que dejar√°n de comprar en los pr√≥ximos 3 meses (90 d√≠as).
# Siguiendo la l√≥gica de churn de la Clase 10:
# - Si un cliente lleva m√°s de 90 d√≠as sin comprar, lo consideramos INACTIVO (0),
#   es decir, en abandono.
# - Si compr√≥ en los √∫ltimos 90 d√≠as, lo consideramos ACTIVO (1).

umbral_dias = 90  # 90 d√≠as ~ 3 meses

df_tasa_abandono["tasa_abandono_label"] = np.where(
    df_tasa_abandono["dias_desde_ultima_compra"] > umbral_dias,
    0,  # inactivo ‚Üí alta probabilidad de abandono
    1   # activo ‚Üí sigue comprando
)

print("\nDistribuci√≥n de la etiqueta de tasa de abandono (0 = inactivo, 1 = activo):")
print(df_tasa_abandono["tasa_abandono_label"].value_counts())

# ------------------------------------------------------------------------------
# 5. Definir X (atributos) e y (etiqueta)
# ------------------------------------------------------------------------------

y_tasa = df_tasa_abandono["tasa_abandono_label"]

features_numericas = ["cant_compras", "monto_promedio", "dias_desde_ultima_compra"]
features_categoricas = ["ciudad", "categoria_mas_comprada", "medio_pago_mas_usado"]

X_tasa = df_tasa_abandono[features_numericas + features_categoricas].copy()

# Manejo simple de nulos
for col in features_numericas:
    X_tasa[col] = X_tasa[col].fillna(0)

for col in features_categoricas:
    X_tasa[col] = X_tasa[col].fillna("DESCONOCIDO")

# One-Hot Encoding para variables categ√≥ricas
X_tasa_encoded = pd.get_dummies(
    X_tasa,
    columns=features_categoricas,
    drop_first=True
)

print("\nShape de X_tasa_encoded:", X_tasa_encoded.shape)

# ------------------------------------------------------------------------------
# 6. Divisi√≥n Train/Test
# ------------------------------------------------------------------------------

X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
    X_tasa_encoded,
    y_tasa,
    test_size=0.2,
    random_state=42,
    stratify=y_tasa
)

print(f"\nTrain clientes: {X_train_t.shape[0]} | Test clientes: {X_test_t.shape[0]}")

# ------------------------------------------------------------------------------
# 7. Entrenamiento del modelo (Regresi√≥n Log√≠stica)
# ------------------------------------------------------------------------------

modelo_tasa = LogisticRegression(max_iter=1000)
modelo_tasa.fit(X_train_t, y_train_t)

print("\nModelo de Regresi√≥n Log√≠stica entrenado correctamente para tasa de abandono.")

# ------------------------------------------------------------------------------
# 8. Evaluaci√≥n del modelo
# ------------------------------------------------------------------------------

y_pred_t = modelo_tasa.predict(X_test_t)

acc_tasa  = accuracy_score(y_test_t, y_pred_t)
prec_tasa = precision_score(y_test_t, y_pred_t, zero_division=0)
rec_tasa  = recall_score(y_test_t, y_pred_t, zero_division=0)
f1_tasa   = f1_score(y_test_t, y_pred_t, zero_division=0)

print("\nM√©tricas del modelo de tasa de abandono:")
print(f"Accuracy : {acc_tasa:.3f}")
print(f"Precision: {prec_tasa:.3f}")
print(f"Recall   : {rec_tasa:.3f}")
print(f"F1-score : {f1_tasa:.3f}")

print("\nReporte de clasificaci√≥n:")
print(classification_report(y_test_t, y_pred_t, digits=3))

# ------------------------------------------------------------------------------
# 9. Dashboard simple de tasa de abandono
# ------------------------------------------------------------------------------

# 9.1 Tabla de m√©tricas
metrics_tasa = pd.DataFrame({
    "Accuracy": [acc_tasa],
    "Precision": [prec_tasa],
    "Recall": [rec_tasa],
    "F1-score": [f1_tasa]
}, index=["Regresi√≥n Log√≠stica - Tasa de abandono"])

print("\nResumen de m√©tricas del modelo de tasa de abandono:")
display(metrics_tasa)

# 9.2 Matriz de confusi√≥n
cm_tasa = confusion_matrix(y_test_t, y_pred_t)

plt.figure(figsize=(5, 4))
sns.heatmap(
    cm_tasa,
    annot=True,
    fmt="d",
    cmap="Greens",
    xticklabels=["Pred. Inactivo (0)", "Pred. Activo (1)"],
    yticklabels=["Real Inactivo (0)", "Real Activo (1)"]
)
plt.title("Matriz de confusi√≥n ‚Äì Modelo de tasa de abandono")
plt.xlabel("Predicci√≥n")
plt.ylabel("Real")
plt.tight_layout()
plt.show()

# 9.3 Importancia aproximada de variables (coeficientes)
coef_tasa = pd.DataFrame({
    "variable": X_tasa_encoded.columns,
    "coeficiente": modelo_tasa.coef_[0]
})
coef_tasa["importancia_abs"] = coef_tasa["coeficiente"].abs()
coef_tasa_top = coef_tasa.sort_values("importancia_abs", ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(data=coef_tasa_top, x="importancia_abs", y="variable")
plt.title("Top 10 variables m√°s influyentes en la tasa de abandono")
plt.xlabel("Importancia (|coeficiente|)")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()

print("\nFase 8 completada: modelo de tasa de abandono entrenado, evaluado y visualizado.")
```
